```python
import pandas as pd
```

```python
df_input=pd.read_csv("./aoc_2003_prod.txt", header=None)
```

```python
#df_input
```

```python
df_pattern=df_input[0].str.split("", expand=True)
df_pattern=df_pattern.drop(columns=[0, len(df_pattern.columns)-1]).T.reset_index(drop=True).T

df_pattern
```

```python
xmax=len(df_pattern.columns)
```

```python
paths1=[
    (3,1)
]
```

```python
total=1
for stepX, stepY in paths1:
    x = 0
    result=0
    for y in range(0, len(df_pattern.index), stepY):
        result += int(df_pattern.iloc[y][x]=='#')
        x= (x+stepX) % xmax  
    total *= result
    print(f"For steps {stepX}, {stepY}: {result}, {total}")
```
